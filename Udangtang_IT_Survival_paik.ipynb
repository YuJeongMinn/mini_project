```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = "https://paikdabang.com/menu/menu_coffee/"

page = urlopen(url)

soup = BeautifulSoup(page, "lxml") # lxml : parser
# soup = BeautifulSoup(page, "html.parser")
```

```python
print(soup.prettify())
```

```python
menu = soup.find_all("div", class_ = "hover")
print(len(menu))
```

```python
raw_data = {"brand" : [      ], "coffee" : [    ], "coffee_name" : [    ],  "price" : [    ], "caffeine" : [    ], "sugar" : [    ], "kcal" : [    ]}
```

```python
for m in menu:
    raw_data["coffee_name"].append(m.h3.string)
```

```python
for m in menu:
    print(m.div)
```

```python
for m in menu:
    ingredient = m.find("div", class_="ingredient_table_box")
    if ingredient.div is None:
        print("--------------------none-------------------------")
        raw_data["caffeine"].append(-1)
    else:
        divs = ingredient.find_all("div")
        print(divs[0])
        print(divs[1])
        raw_data["caffeine"].append(divs[1].string)
```

```python
print(raw_data)
```

```python
for m in menu:
    ingredient = m.find("div", class_="ingredient_table_box")
    if ingredient.div is None:
        print("--------------------none-------------------------")
        raw_data["kcal"].append(-1)
    else:
        divs = ingredient.find_all("div")
        print(divs[2])
        print(divs[3])
        raw_data["kcal"].append(divs[3].string)
```

```python
print(raw_data)
```

```python
for m in menu:
    ingredient = m.find("div", class_="ingredient_table_box")
    if ingredient.div is None:
        print("--------------------none-------------------------")
        raw_data["sugar"].append(-1)
    else:
        divs = ingredient.find_all("div")
        print(divs[6])
        print(divs[7])
        raw_data["sugar"].append(divs[7].string)
```
